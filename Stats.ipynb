```python
def cal_variance(data):
  n=len(data)
  mean=sum(data)/n

  # calculate the deviation

  deviation=[(x-mean)**2 for x in data]

  # calculate the variance
  variance=sum(deviation)/n

  return variance


cal_variance(data_copy)
```
