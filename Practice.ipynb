```python
# loops are using in used defiened function

def print_even_number():
  n = int(input("Enter the number: "))

  even_number = []
  for num in range(1, n + 1):
    if num % 2 == 0:
      even_number.append(num)
  print("Even number up to", n, "are", even_number)

print_even_number()
```
