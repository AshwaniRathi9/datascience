```python
def hello_decorator(func):
  def inner():
    print("before execution")
    value=func(a,b)
    print("after execution")
    return value
  return inner
```

```python
@hello_decorator
def sum_two_number(a,b):
  print("inside this function")
  return a+b
```

```python
sum_two_number(10,20)
```
